```python
import numpy as np
import pandas as pd

train_csv = pd.read_csv('/root/train.csv')
test_csv = pd.read_csv('/root/test.csv')
```

```python
mental_train=pd.DataFrame(train_csv)
mental_test=pd.DataFrame(test_csv)
```

```python
# dropping name,city and degree since we don't need those
mental_train.drop(['Name','City','Degree'],axis=1,inplace=True)
mental_test.drop(['Name','City','Degree'],axis=1,inplace=True)
```

```python
mental_train["Work/Academic Pressure"] = mental_train.apply(
    lambda row: row['Academic Pressure'] if row["Working Professional or Student"] == "Student" else row["Work Pressure"], axis=1
)
mental_test["Work/Academic Pressure"] = mental_test.apply(
    lambda row: row['Academic Pressure'] if row["Working Professional or Student"] == "Student" else row["Work Pressure"], axis=1
)
mental_train["Job/Study Satisfaction"] = mental_train.apply(
    lambda row: row['Study Satisfaction'] if row["Working Professional or Student"] == "Student" else row["Job Satisfaction"],axis=1
)
mental_test["Job/Study Satisfaction"] = mental_test.apply(
    lambda row: row['Study Satisfaction'] if row["Working Professional or Student"] == "Student" else row["Job Satisfaction"],axis=1
)
```

```python
mental_train.drop(['Academic Pressure','Work Pressure','Study Satisfaction','Job Satisfaction'],axis=1,inplace=True)
mental_test.drop(['Academic Pressure','Work Pressure','Study Satisfaction','Job Satisfaction'],axis=1,inplace=True)
```

```python
mental_train.loc[mental_train.groupby('Dietary Habits')['Dietary Habits'].transform('count') < 10,'Dietary Habits'] = np.nan
mental_test.loc[mental_test.groupby('Dietary Habits')['Dietary Habits'].transform('count') < 10,'Dietary Habits'] = np.nan
```

```python
mental_train['Dietary Habits'].fillna(mental_train['Dietary Habits'].mode()[0], inplace=True)
mental_test['Dietary Habits'].fillna(mental_test['Dietary Habits'].mode()[0], inplace=True)
```

```python
mental_train.loc[(mental_train['Profession'].isnull()) & (mental_train['Working Professional or Student'] == 'Working Professional'), 'Profession'] = 'Unknown'
mental_test.loc[(mental_test['Profession'].isnull()) & (mental_test['Working Professional or Student'] == 'Working Professional'), 'Profession'] = 'Unknown'
```

```python
mental_train.loc[(mental_train.groupby('Profession')['Profession'].transform('count') < 10) &
                 (mental_train['Working Professional or Student'] == 'Working Professional'),'Profession'] = 'Unknown'

mental_test.loc[(mental_test.groupby('Profession')['Profession'].transform('count') < 10) &
                 (mental_test['Working Professional or Student'] == 'Working Professional'),'Profession'] = 'Unknown'
```

```python
student_df = mental_train[mental_train['Working Professional or Student'] == 'Student']

# Calculate the sum of null values in the 'Profession' column for students
sum_null_profession_student = student_df['Profession'].isnull().sum()

print(f"The sum of null values in the 'Profession' column for students is: {sum_null_profession_student}")
```

```python
# prompt: mental_train['CGPA'].fillna(mental_train['CGPA'].mean(), inplace=True)   for those where working professional or student == student

student_df = mental_train[mental_train['Working Professional or Student'] == 'Student']
student_df['CGPA'].fillna(student_df['CGPA'].mean(), inplace=True)
mental_train.loc[mental_train['Working Professional or Student'] == 'Student', 'CGPA'] = student_df['CGPA']
mental_test.loc[mental_test['Working Professional or Student'] == 'Student', 'CGPA'] = student_df['CGPA']
```

```python
mental_train['Work/Academic Pressure'].fillna(mental_train['Work/Academic Pressure'].mode()[0], inplace=True)
mental_test['Work/Academic Pressure'].fillna(mental_test['Work/Academic Pressure'].mode()[0], inplace=True)
```

```python
mental_train['Job/Study Satisfaction'].fillna(mental_train['Job/Study Satisfaction'].mode()[0], inplace=True)
mental_test['Job/Study Satisfaction'].fillna(mental_test['Job/Study Satisfaction'].mode()[0], inplace=True)
```

```python
mental_train['Financial Stress'].fillna(mental_train['Financial Stress'].mode()[0], inplace=True)
mental_test['Financial Stress'].fillna(mental_test['Financial Stress'].mode()[0], inplace=True)
```

```python
mental_train['Sleep Duration'] = mental_train['Sleep Duration'].apply(lambda row: "More than 7 hours" if row in ['More than 8 hours','7-8 hours','8-9 hours'] else'Less than 7 hours')
mental_train['Sleep Duration'].value_counts()
mental_test['Sleep Duration'] = mental_test['Sleep Duration'].apply(lambda row: "More than 7 hours" if row in ['More than 8 hours','7-8 hours','8-9 hours'] else'Less than 7 hours')
```

```python

columns_to_change = ['Age', 'Work/Study Hours', 'Financial Stress', 'Work/Academic Pressure', 'Job/Study Satisfaction']
for column in columns_to_change:
      mental_train[column] = mental_train[column].astype(int)
      mental_test[column] = mental_test[column].astype(int)
```

```python
X=mental_train.drop(['Depression','id'],axis='columns')
y=mental_train['Depression']
```

```python
cat_binary_features = X.select_dtypes(include=['object']).columns[
    X.select_dtypes(include=['object']).nunique() == 2
]
cat_multi_features = X.select_dtypes(include=['object']).columns[
    mental_train.select_dtypes(include=['object']).nunique() > 2
]
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
numeric_transformer=StandardScaler()
oh_transformer=OneHotEncoder(drop='first', handle_unknown='ignore') # Added handle_unknown='ignore'
```

```python
preprocessor=ColumnTransformer(
    [
     ('OneHotEncoder',oh_transformer,cat_multi_features),
     ('StandardScaler',numeric_transformer,num_features)
    ]
)
```

```python

# from sklearn.preprocessing import LabelEncoder

# columns_to_encode = ['Gender','Working Professional or Student','Sleep Duration', 'Have you ever had suicidal thoughts ?','Family History of Mental Illness']
# le = LabelEncoder()
# for col in columns_to_encode:
#     X[col] = le.fit_transform(X[col])
#     mental_test[col] = le.transform(mental_test[col])

import joblib
from sklearn.preprocessing import LabelEncoder

# Define columns and categories for encoding
columns_to_encode = {
    'Gender': ['Male', 'Female'],
    'Working Professional or Student': ['Working Professional', 'Student'],
    'Sleep Duration': ['More than 7 hours', 'Less than 7 hours'],
    'Have you ever had suicidal thoughts ?': ['Yes', 'No'],
    'Family History of Mental Illness': ['Yes', 'No']
}

le_dict = {}

# Fit the LabelEncoder on each column and save them
for col, classes in columns_to_encode.items():
    le = LabelEncoder()
    le.fit(classes)  # Fit the LabelEncoder on known classes

    # Transform the training data
    X[col] = le.transform(X[col])
    
    # Sanitize the column name for the filename (replace spaces with underscores, remove special characters)
    safe_filename = col.replace(" ", "_").replace("?", "") + "_le.pkl"
    
    # Save the encoder
    joblib.dump(le, safe_filename)  # Save encoder for future use
    
    # Store the encoder in the dictionary for later use
    le_dict[col] = le
    
    # Transform the test data using the saved encoder
    mental_test[col] = le_dict[col].transform(mental_test[col])

```

```python
X=preprocessor.fit_transform(X)
mental_test=preprocessor.transform(mental_test)
```
